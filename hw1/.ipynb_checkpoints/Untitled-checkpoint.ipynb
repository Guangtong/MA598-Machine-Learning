```python
# import data

import pandas as pd
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/pima-indians-diabetes/pima-indians-diabetes.data'
diabetes = pd.read_csv(url, header=None)
X=diabetes.drop(8, axis=1)
classification = diabetes[8].values
```

# K-means clustering

```python
# K-means with K clusters
K = 2
from sklearn.cluster import KMeans
km = KMeans(n_clusters=K, random_state=1)
km.fit(X)
# save the cluster centera
center_df = pd.DataFrame(km.cluster_centers_)
```

```python
total = X.append(center_df, ignore_index = True)

from sklearn.decomposition import PCA
pca= PCA(n_components = 2)
total_pca = pca.fit_transform(total)
X_pca= total_pca[0:767]
center_pca = total_pca[768:769]
```

```python
# create a "colors" array for plotting
import numpy as np
colors = np.array(['red', 'green', 'blue', 'yellow'])

# allow plots to appear in the notebook
%matplotlib inline
import matplotlib.pyplot as plt

# scatter plot of attribut 1 versus 2, colored by cluster (0=red, 1=green, 2=blue)
plt.scatter(X_pca.T[0], X_pca.T[1], c=colors[km.labels_], linewidths=0, s=50)

# cluster centers, marked by "."
plt.scatter(center_pca.T[0], center_pca.T[1], linewidths=3,s=100, c='white')

# add labels
plt.xlabel('x0')
plt.ylabel('x1')
```
